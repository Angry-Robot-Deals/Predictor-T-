```python
import os
import json
import pandas as pd
```

```python
data = pd.DataFrame(columns=["time", "o", "c", "h", "l", "v"])
```

```python
directory = 'BNBBUSD/1min'
```

```python
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        with open(os.path.join(directory, filename), 'r') as file:
            file_content = file.read()
            try:
                json_data = json.loads(file_content)
                data = pd.concat([data, pd.DataFrame(json_data)], ignore_index=True)
            except json.JSONDecodeError as e:
                print(f"Error reading {filename}: {e}")
```

```python
data
```

```python
# 1. Extract OHLC data and create a new DataFrame
new_data = pd.DataFrame()
new_data['Open'] = data['o']
new_data['High'] = data['h']
new_data['Low'] = data['l']
new_data['Close'] = data['c']

# 2. Save the new DataFrame to a .csv file
new_data.to_csv('one_minute.csv', index=False)
```
